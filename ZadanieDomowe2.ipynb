```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab2-Transactions")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} — gotowy")
```

```python
df = spark.read.json("transactions_10k.jsonl")

print(f"Liczba rekordów: {df.count()}")
df.printSchema()
```

```python
df.show(10, truncate=False)
```

```python
from pyspark.sql.functions import to_timestamp, col

df = df.withColumn("timestamp", to_timestamp(col("timestamp"), "yyyy-MM-dd HH:mm:ss"))

df.printSchema()  # timestamp powinien być teraz 'timestamp (nullable = true)'
```

```python
from pyspark.sql.functions import count, sum as _sum, avg, round as _round

store_summary = (
    df.groupBy("store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _round(avg("amount"), 2).alias("srednia_PLN"),
    )
    .orderBy("store")
)
store_summary.show()
```

```python
from pyspark.sql.functions import min as _min, max as _max

category_stats = (
    df.groupBy("category")
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN"),
        _min("amount").alias("min_PLN"),
        _max("amount").alias("max_PLN"),
        count("tx_id").alias("liczba_transakcji")
    )
    .orderBy("category")
)

category_stats.show()
```

```python
from pyspark.sql.functions import window

hourly = (
    df.groupBy(window("timestamp", "1 hour"))    # okno 1-godzinne
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .orderBy("window")
)
hourly.show(truncate=False)
```

```python
(
    hourly
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .show(truncate=False)
)
```

```python
store_30min = (
    df.groupBy(window("timestamp", "30 minutes"), "store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy("window", "store")
)

store_30min.show(truncate=False)
```

```python
from pyspark.sql.functions import desc

krakow_peak_hour = (
    df.filter(col("store") == "Kraków")
    .groupBy(window("timestamp", "1 hour"))
    .agg(
        _round(_sum("amount"), 2).alias("suma_PLN")
    )
    .orderBy(desc("suma_PLN"))
)

krakow_peak_hour.show(truncate=False)
```

```python
sliding = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))  # szerokość 1h, krok 30min
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
    .select(
        col("window.start").alias("od"),
        col("window.end").alias("do"),
        "liczba_tx",
        "suma_PLN",
    )
    .orderBy("od")
)
sliding.show(truncate=False)
```

```python
tumbling_rows = (
    df.groupBy(window("timestamp", "1 hour"))
    .agg(count("tx_id"))
    .count()
)

sliding_rows = (
    df.groupBy(window("timestamp", "1 hour", "30 minutes"))
    .agg(count("tx_id"))
    .count()
)

print(f"Tumbling (1h):          {tumbling_rows} okien")
print(f"Sliding  (1h / 30min):  {sliding_rows} okien")

# ODPOWIEDŹ: 
# Sliding window generuje więcej wierszy, ponieważ okna nakładają się na siebie. 
# W tym przypadku nowe okno tworzy się co 30 minut, podczas gdy każde z nich trwa pełną godzinę.
```

```python
# 1. Ile transakcji jest w oknie 09:00–10:00?
#    Sprawdź w wyniku zadania 3.1.
#    ODPOWIEDŹ: W oknie 09:00–10:00 znajduje się 4661 transakcji.

# 2. Jaka jest różnica między groupBy("store") a groupBy(window(...), "store")?
#    groupBy("store") wykonuje agregację globalną dla każdego sklepu (wynikiem 
#    jest jeden wiersz na sklep za cały okres danych). 
#    groupBy(window(...), "store") dzieli dane na konkretne przedziały czasowe, 
#    pozwalając analizować dynamikę sprzedaży i np. wyznaczyć godziny szczytu dla każdego sklepu.

# 3. W oknie sliding 1h/30min — ile okien zawiera transakcje z godziny 09:30?
#    Wskazówka: narysuj oś czasu.
#    ODPOWIEDŹ: Transakcja z godziny 09:30 wpada do 2 okien
```

```python
from pyspark.sql.functions import asc

gdansk_min_avg = (
    df.filter(col("store") == "Gdańsk")
    .groupBy(window("timestamp", "1 hour"))
    .agg(_round(avg("amount"), 2).alias("srednia_kwota"))
    .orderBy(asc("srednia_kwota"))
    .select("window.start", "window.end", "srednia_kwota")
)

gdansk_min_avg.show(1, truncate=False)
```

```python
from pyspark.sql.functions import col, count

df.filter(
    (col("timestamp") >= "2026-04-12 09:00:00") &
    (col("timestamp") <  "2026-04-12 09:30:00")
).groupBy("category").agg(
    count("tx_id").alias("liczba_tx")
).orderBy("category").show()
```

```python
peak_quarter = (
    df.groupBy(window("timestamp", "15 minutes"))
    .agg(count("tx_id").alias("liczba_transakcji"))
    .orderBy(desc("liczba_transakcji"))
    .select("window.start", "window.end", "liczba_transakcji")
)

peak_quarter.show(truncate=False)

# ODPOWIEDŹ: 09:15 do 09:30
```
